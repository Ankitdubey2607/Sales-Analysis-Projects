```python
# Assuming df is your DataFrame
df['Total Amount'] = df['UnitPrice'] * df['Quantity'] - df['DiscountAmount']
```

```python
df = pd.DataFrame(df)
```

```python
df['idCustomer'].nunique()
```

```python
country_amount=df.groupby(['Country'], as_index=False)['Total Amount'].sum().sort_values(by='Total Amount', ascending=False).head(10)
```

```python
sns.set(rc={'figure.figsize':(12,8)})
sns.barplot(x="Country", y='Total Amount', data=country_amount)
```

```python
grouped_df=df.groupby(['Country', 'Product'], as_index=False)['Total Amount'].sum().sort_values(by='Total Amount', ascending=False)
top_10_product=grouped_df.groupby('Product')['Total Amount'].sum().nlargest(10).index
top_10_country=grouped_df.groupby('Country')['Total Amount'].sum().nlargest(10).index
filtered_df=grouped_df[(grouped_df['Product'].isin(top_10_product))&(grouped_df['Country'].isin(top_10_country))]
filtered_df=filtered_df.sort_values(by=['Product', 'Total Amount'], ascending=[True, False])
```

```python
sns.barplot(x='Country', y='Total Amount', data=filtered_df)
```

 The top 10 products in terms of total amount, categorized by the top 10 countries with the maximum total amount.

```python
sns.set(rc={'figure.figsize':(16,8)})
sns.barplot(x='Product', y='Total Amount', data=filtered_df)
```

```python
df.head()
```

```python
Brand_df=df.groupby(['Brand'], as_index=False)['Total Amount'].sum().sort_values(by='Total Amount', ascending=False)
sns.barplot(x='Brand', y='Total Amount', data=Brand_df)
```

Adventure Works has a maximum Total Amount and Northwind Traders has minimum Total Amount.

```python
Brand_unitprice=df.groupby(['Brand'], as_index=False)['UnitPrice'].min().sort_values(by='UnitPrice', ascending=False)
sns.barplot(x='Brand', y='UnitPrice', data=Brand_unitprice)
```

So Brand Fabrikam has a highest unit price while Contoso has least unit price.

```python
Brand_quantity=df.groupby(['Brand'], as_index=False)['Quantity'].sum().sort_values(by='Quantity', ascending=False)
sns.barplot(x='Brand', y='Quantity', data=Brand_quantity)
```

Adventure Works has ordered maximum quantity & Litware ordered minimum quantity.

```python
# Group by 'Brand' and 'City', then count occurrences
grouped = df.groupby(['Brand', 'City']).size().reset_index(name='Count')
top_cities_per_brand = grouped.sort_values(['Brand', 'Count'], ascending=[True, False]).groupby('Brand').head(10)

print(top_cities_per_brand)
```

```python
sns.barplot(x='City', y='Count', data=
            top_cities_per_brand[top_cities_per_brand['Brand']== 'Adventure Works'])
plt.title('Top 10 City Brand Adventure Works')
```
